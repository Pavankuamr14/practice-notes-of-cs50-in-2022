```python
def find_winner(arr, k):
    win_count = {num: 0 for num in arr}

    while True:
        if win_count[arr[0]] == k:
            return arr[0]
        if win_count[arr[1]] == k:
            return arr[1]

        winner = max(arr[0], arr[1])
        loser = min(arr[0], arr[1])
        win_count[winner] += 1
        arr.remove(loser)
        arr.append(loser)

# Example 1
arr1 = [2, 1, 3, 5, 4, 6, 7]
k1 = 2
result1 = find_winner(arr1, k1)
print(f"Example 1: {result1}")

# Example 2
arr2 = [3, 2, 1]
k2 = 10
result2 = find_winner(arr2, k2)
print(f"Example 2: {result2}")
```
